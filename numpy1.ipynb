```python
#4*4 matrix
z=np.array([[1,2,3,4],[2,3,4,5],[8,9,7,6],[1,6,7,8]])
z
```

```python
#indexing
print(z[2,2])
print(z[3,1])
print(z[1,2:])
print(z[1:3,1:3])
print(z[0:3,1:3]) #0:3 means no of rows and 1:3 means no of columns
print(z[1:,1:])
```

```python
#create two 2*2 matrix
a=np.array([[1,2],[2,3]])
b=np.array([[5,6],[8,9]])
print(a+b)
print(a-b)
print(a*b) #not a matrix multiplication
#to multiply 2 matrix use a@b or a.dot(b)
print(a/b)
```
